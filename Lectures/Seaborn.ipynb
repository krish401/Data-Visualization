```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt 
```

```python
data = pd.read_csv("Data/mpg.csv")
```

```python
data
```

```python
data = sns.load_dataset("mpg")
data
```

```python
cond = data["model_year"]==70
data[cond].head()
```

```python
'''
errorbar
palette = "dark:#589_r" "pastel" 
color
width
'''

plt.figure(figsize=(12,6))
sns.barplot(data, x="model_year", y="displacement", color="Blue", errorbar =None, width= 0.2)

plt.xlabel("Model Year")
plt.ylabel("Displacement")
plt.title("Bar chart Displacement")
```

```python
'''
create a barplot for displacement against cyliners
.set()
'''

plt.figure(figsize= (10,4))
sns.barplot(data, x = "model_year", y = "displacement").set(
    title = "Bar chart Displacement",
    xlabel = "Model Year",
    ylabel = "Displacement"
)
```

```python
''' 
scatterplot
size = "cylinder"  the size of the points are based on the number of cylinders
hue = "origin"   color change for the origin
style = origin
'''

plt.figure(figsize=(6,6))
sns.scatterplot(data, x= "weight", y = "horsepower", size="cylinders", hue="origin")

plt.xlabel("Weight")
plt.ylabel("Horsepower")
plt.title("Scatterplot Vehicles")
plt.show()
```

```python
''' 
sp.get_legend_handles_labels()
'''

plt.figure(figsize=(6,6))
sp = sns.scatterplot(data, x= "weight", y="horsepower", hue = 'origin')
handles, labels = sp.get_legend_handles_labels()
plt.legend(handles, labels, title="Origin")

plt.xlabel("Weight")
plt.ylabel("Horsepower")
plt.title("Scatterplot Vehicals")
plt.show()
```

```python
'''
Create two line plots (model_year - acceleration and model_year - cylinders) in one graph
'''

gig, ax1 = plt.subplots(figsize=(12,4))
sns.lineplot(data, x="model_year", y='acceleration')
ax1.set_xlabel("Model Year")
ax1.set_ylabel("Acceleration")

ax2 =ax1.twinx()
sns.lineplot(data, x="model_year", y="cylinders")
ax2.set_ylabel("Cylinders")

plt.show()
```

```python
fig, axes = plt.subplots(2,2, figsize=(12,6))
sns.lineplot(data, x= "model_year", y="acceleration", ax = axes[0][0])
axes[0][0].set_title("Acceleration Example")

sns.lineplot(data, x="model_year", y="cylinders", ax=axes[0][1])
axes[0][1].set_title("Cylinders Example")

sns.lineplot(data, x="model_year", y="horsepower", ax= axes[1][0])
axes[1][0].set_title("Horsepower Example")

sns.lineplot(data, x="model_year", y="displacement", ax= axes[1][1])
axes[1][1].set_title("Displacement Example")

plt.tight_layout()
```
