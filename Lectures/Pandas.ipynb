```python
'''
Find mean Value for Each City
Find the size of each category
Display the sum of Value after grouping by City and Category

df = pd.DataFrame({'City': ['Stratford', 'Toronto', 'Stratford', 'Toronto', 'Kitchener'],
        'Category': ['A', 'B', 'A', 'A', 'B'],
        'Value': [10, 20, 15, 30, 25]}
                 )
'''

df = pd.DataFrame({'City': ['Stratford', 'Toronto', 'Stratford', 'Toronto', 'Kitchener'],
        'Category': ['A', 'B', 'A', 'A', 'B'],
        'Value': [10, 20, 15, 30, 25]}
                 )

df2 = df.groupby("City")["Value"].mean()
print(df2)
df3 = df.groupby("Category").size()
print(df3)
df4 = df.groupby(["Category", "Value"]).sum()
print(df4)
```
