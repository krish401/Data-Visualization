```python
x = np.linspace(0,10,100) #100 elements evenly spaced between 0 and 10
y1 = np.sin(x)
y2 = np.cos(x)

plt.plot(x, y1, label ="Sin Wave", color = "green", linestyle = '-' )
plt.plot(x,y2, label="Cos graph", color="blue", linestyle='--')

plt.xlabel("Values")
plt.ylabel("f(x)")
plt.legend()
plt.grid(True)
plt.show()
```

```python
data = [np.random.normal(0,1,100),
        np.random.normal(2,1.5,150),
        np.random.normal(-1, 0.5, 75)
        ]

label = ["Group 1", "Group 2", "Group 3"]

plt.figure(figsize=(8,5))
plt.boxplot(data, labels= label, showfliers=False) # hides the outliers

plt.xlabel("Groups")
plt.ylabel("Values")
plt.show()
```

```python
np.random.seed(42)

data1 = np.random.normal(0,1, 1000)
data2 = np.random.normal(2, 1.5, 1000)

plt.figure(figsize=(8,5))
plt.hist(data1, label="data 1", bins = 20, color="green", alpha = 0.7, edgecolor= "black")
plt.hist(data2, label="data 2", bins = 20, color="red", alpha = 0.7, edgecolor= "black")
plt.xlabel("Values")
plt.ylabel("Frequency")

plt.legend()
plt.title("Histogram EX")
plt.show()
```

```python
import yfinance as yf
```

```python
def download_stock_data(ticker, date, timeframe):
    start_date = pd.to_datetime(date)
    end_date = start_date + pd.Timedelta(days= 1)

    data = yf.download(ticker,start=start_date, end=end_date, interval=timeframe)

    return data

def plot_stock_data(data, ticker):
    fig, ax1 = plt.subplots(figsize=(12,6))
    ax1.plot(data.index, data['Close'], color = "blue", label="Close")
    ax1.set_xlabel("Time")
    ax1.set_ylabel("Price")

    ax2 = ax1.twinx()
    ax2.plot(data.index, data["Volume"], color= 'gray', label= 'Volume', alpha = 0.4)
    ax2.set_ylabel("Volume")

    plt.title("Stock Data")
    fig.legend()

data = download_stock_data("GOOG", "2025-02-06,", "5m")
plot_stock_data(data, "GOOG")
data.head()

```
