```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# All three function will be here

data = pd.read_csv('data_assignment4.csv')

data.set_index("Geography", inplace=True)

data_transpose = data.T

data_transpose.index = pd.date_range(start="2023-01", periods=len(data_transpose), freq="M")

def plot_toronto_vs_canada():
    plt.figure(figsize=(10,5))
    plt.plot(data_transpose.index, data_transpose.loc[:, "Toronto, Ontario"], label="Toronto", marker='o')
    plt.plot(data_transpose.index, data_transpose.loc[:, "Canada"], label="Canadian Average", linestyle='--', marker='s')

    plt.xlabel("Time")
    plt.ylabel("Price Index")
    plt.title("Toronto vs Canada Average Price Trends")
    plt.legend()
    plt.grid(True)
    
    plt.show()

def plot_price_volatility():
    std_dev = data_transpose.std().sort_values(ascending=False)
    highest_volatility = std_dev.head(3)
    lowest_volatility = std_dev.tail(3)

    plt.figure(figsize=(10,5))
    plt.bar(highest_volatility.index, highest_volatility.values, color= 'red', label="Higest Volatility")
    plt.bar(lowest_volatility.index, lowest_volatility.values, color='green', label="lowest Volatility")

    plt.xlabel("Region")
    plt.ylabel("Standard Deviation of Prices")
    plt.title("Price Volatility Across Regions")
    plt.legend()
    plt.xticks(rotation=45, ha="right")
    
    plt.show

def plot_toronto_vs_ontario():
    plt.figure(figsize=(10,5))
    plt.plot(data_transpose.index, data_transpose.loc[:, "Toronto, Ontario"], label= "Toronto", marker='o')
    plt.plot(data_transpose.index, data_transpose.loc[:, "Ottawa-Gatineau, Ontario part, Ontario/Quebec 2"], label="Ottawa", linestyle='--', marker='s')
    plt.plot(data_transpose.index, data_transpose.loc[:, "Thunder Bay, Ontario 3"], label="Thunder Bay", linestyle=":", marker="^")

    plt.xlabel("Time")
    plt.ylabel("Price Index")
    plt.title("Toronto vs. Ottawa vs. Thunder Bay Price Trends")
    plt.legend()
    plt.grid(True)

    plt.show()

plot_toronto_vs_canada()
plot_price_volatility()
plot_toronto_vs_ontario()

```
