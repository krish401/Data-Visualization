# Assignment 2 Pandas

```python
import pandas as pd
```

## 2. Display Head

```python
df = pd.read_csv('Assignment2Data.csv')
df.head()
```

## 3. 5 Random Rows

```python
df.sample(5)
```

## 4. Wins > Match 

```python
cond = df["Won"] > df["Matches"]
df.loc[cond]
```

## 5. Delete Row From 4

```python
df = df.drop(df.loc[cond].index)
df

# you can also simply just get the index and drop from the index
# but i chose to use the condition
```

## 6. Create Column Lost

```python
df["Lost"] = df["Matches"] - df["Won"]
df
```

## 7. Create Win Percent

```python
df["Win percentage"] = df["Won"]/df["Matches"]
df
```

## 8. Create Category

```python
def category(t):

    if t>.9:
        return "Professional"
    elif t>.75:
        return "Target"
    else:
        return "Development"
    

df["Category"] = df["Win percentage"].map(category)
df
```

## 9. Display Player ID of Highest Win

```python
sorted_df = df.sort_values("Win percentage", ascending=False)
higest_rated = sorted_df.iloc[0]["Player ID"]
print(higest_rated)
```

## 10. Display Player ID of Lowest Win

```python
lowest_rated = sorted_df.iloc[-1]["Player ID"]
print(lowest_rated)
```

## 11. Group based on category

```python
best_players = df.loc[df.groupby("Category")["Win percentage"].idxmax()]
best_players
```

## 12. List which age is doing best in Waterloo

```python
df_waterloo = df.loc[df["City"] == "Waterloo"]

def age(age):
    if age <= 20:
        return "20 and below"
    elif 21 <= age <= 25:
        return "21-25"
    elif 26 <= age <= 30:
        return "26-30"
    else:
        return "30 and above"
    
df_waterloo["Age group"] = df_waterloo["Age"].map(age)

best_age_group = df_waterloo.groupby("Age group")["Win percentage"].mean().idxmax()
print(f'The best age group in waterloo is: {best_age_group}')
```

## 13. Find win for each city

```python
city_win_percent = df.groupby("City")["Win percentage"].mean()
city_win_percent
```

## 14. Combined win percentage over 20

```python
df_above_20 = df.loc[df["Age"] > 20]
win_percent = ((df_above_20["Won"]).sum() / df_above_20["Matches"].sum()) *100
print(f'The combined win percentage for players above 20: {win_percent}')
```
